```python
import pandas as pd
import numpy as np
import random
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from annoy import AnnoyIndex
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')

# Setup random seed
RANDOM_SEED = 24
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

# Load dataset
df = pd.read_csv('/content/heart.csv')

print("="*80)
print("ANALISIS CLUSTERING DATASET PENYAKIT JANTUNG")
print("="*80)

# 1. DESKRIPSI DATASET
print("\n1. DESKRIPSI DATASET")
print("-" * 50)
print(f"Jumlah sampel: {df.shape[0]}")
print(f"Jumlah fitur: {df.shape[1]}")
print("\n5 sampel pertama:")
display(df.head())
print("\nRingkasan statistik:")
display(df.describe())

# 2. PREPROCESSING
print("\n2. PREPROCESSING DATA")
print("-" * 50)

# 2.1 Pengecekan Missing Values
print("Missing values per kolom:")
print(df.isnull().sum())

# 2.2 Pembuatan Fitur Baru
df['chol_age'] = df['chol'] * df['age']
print(f"\nFitur baru 'chol_age' berhasil dibuat")
print(f"Statistik chol_age: min={df['chol_age'].min():.1f}, max={df['chol_age'].max():.1f}, mean={df['chol_age'].mean():.1f}")

# 2.3 Normalisasi Data
num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak", "chol_age"]
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
print(f"\nNormalisasi berhasil pada {len(num_cols)} fitur numerik")

# 3. ANALISIS CLUSTERING
print("\n3. ANALISIS CLUSTERING")
print("-" * 50)

# 3.1 Menentukan Jumlah Cluster Optimal (K-Means)
print("3.1 Menentukan Jumlah Cluster Optimal")
wcss = []
s_scores = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=RANDOM_SEED)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

    if k > 1:
        labels = kmeans.fit_predict(df)
        score = silhouette_score(df, labels)
        s_scores.append(score)
        print(f"k={k}: WCSS={kmeans.inertia_:.2f}, Silhouette={score:.4f}")

# Visualisasi Elbow dan Silhouette
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(K_range, wcss, 'bo-', markersize=6)
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('WCSS (Inertia)')
plt.title('Metode Elbow untuk K Optimal')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), s_scores, 'ro-', markersize=6)
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.title('Metode Silhouette untuk K Optimal')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 3.2 Implementasi K-Means dengan k=2
print("\n3.2 Implementasi K-Means (k=2)")
kmeans = KMeans(n_clusters=2, random_state=RANDOM_SEED)
kmeans_labels = kmeans.fit_predict(df)
sil_kmeans = silhouette_score(df, kmeans_labels)
dbi_kmeans = davies_bouldin_score(df, kmeans_labels)
print(f"K-Means berhasil dengan 2 cluster")
print(f"Silhouette Score: {sil_kmeans:.4f}")
print(f"Davies-Bouldin Index: {dbi_kmeans:.4f}")

# 3.3 DBSCAN Clustering
print("\n3.3 Implementasi DBSCAN")

# Menentukan epsilon optimal
neigh = NearestNeighbors(n_neighbors=5)
nbrs = neigh.fit(df)
distances, indices = nbrs.kneighbors(df)
distances = np.sort(distances[:, 4])

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(distances)
plt.ylabel("Jarak Tetangga ke-5 Terdekat")
plt.xlabel("Points diurutkan berdasarkan jarak")
plt.title("Penentuan Epsilon untuk DBSCAN")
plt.grid(True, alpha=0.3)

# Implementasi DBSCAN
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan_labels = dbscan.fit_predict(df)

plt.subplot(1, 2, 2)
unique, counts = np.unique(dbscan_labels, return_counts=True)
plt.bar(unique, counts, color=['red' if x == -1 else 'blue' for x in unique])
plt.xlabel('Label Cluster')
plt.ylabel('Jumlah Points')
plt.title('Distribusi Cluster DBSCAN')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
n_noise = list(dbscan_labels).count(-1)
print(f"DBSCAN menghasilkan {n_clusters_dbscan} cluster")
print(f"Jumlah noise points: {n_noise}")

if n_clusters_dbscan > 1:
    sil_dbscan = silhouette_score(df, dbscan_labels)
    dbi_dbscan = davies_bouldin_score(df, dbscan_labels)
    print(f"Silhouette Score: {sil_dbscan:.4f}")
    print(f"Davies-Bouldin Index: {dbi_dbscan:.4f}")
else:
    sil_dbscan = float('nan')
    dbi_dbscan = float('nan')
    print("Tidak dapat menghitung metrics untuk DBSCAN (hanya 1 cluster)")

# 3.4 Visualisasi Hasil Clustering
print("\n3.4 Visualisasi Hasil Clustering")
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df)

plt.figure(figsize=(12, 5))

# K-Means Visualization
plt.subplot(1, 2, 1)
scatter1 = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=kmeans_labels,
                      cmap='viridis', alpha=0.7, s=30)
plt.colorbar(scatter1, label='Cluster')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-Means Clustering (k=2)')
plt.grid(True, alpha=0.3)

# DBSCAN Visualization
plt.subplot(1, 2, 2)
scatter2 = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=dbscan_labels,
                      cmap='Set3', alpha=0.7, s=30)
plt.colorbar(scatter2, label='Cluster')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('DBSCAN Clustering')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 3.5 Perbandingan Hasil Clustering
print("\n3.5 PERBANDINGAN HASIL CLUSTERING")
print("="*60)
results = pd.DataFrame({
    'Algorithm': ['K-Means', 'DBSCAN'],
    'Silhouette Score': [sil_kmeans, sil_dbscan],
    'Davies-Bouldin Index': [dbi_kmeans, dbi_dbscan],
    'Number of Clusters': [len(set(kmeans_labels)), n_clusters_dbscan],
    'Noise Points': [0, n_noise]
})
display(results.round(3))

# 4. IMPLEMENTASI ANNOY
print("\n4. IMPLEMENTASI ANNOY UNTUK NEAREST NEIGHBORS")
print("-" * 50)

# 4.1 Membangun Annoy Index
f = df.shape[1]
annoy_index = AnnoyIndex(f, 'euclidean')

for i, vector in enumerate(df.values):
    annoy_index.add_item(i, vector)

annoy_index.build(10)
print("Annoy index berhasil dibangun dengan 10 trees")

# 4.2 Query Nearest Neighbors
np.random.seed(RANDOM_SEED)
query_indices = np.random.choice(len(df), 3, replace=False)

print("\nHASIL QUERY NEAREST NEIGHBORS")
print("="*70)

for i, query_idx in enumerate(query_indices, 1):
    k = 5
    neighbors, distances = annoy_index.get_nns_by_item(query_idx, k, include_distances=True)

    print(f"\nQUERY {i}:")
    print(f"Index Query: {query_idx}")
    query_data = df.iloc[query_idx]
    print(f"Fitur penting: age={query_data['age']:.2f}, trestbps={query_data['trestbps']:.2f}, "
          f"chol={query_data['chol']:.2f}, thalach={query_data['thalach']:.2f}")
    print(f"Cluster: {kmeans_labels[query_idx]}")

    print(f"\n{'Rank':<4} {'Index':<8} {'Distance':<12} {'Cluster':<8} {'Age':<8} {'Chol':<8}")
    print("-" * 60)

    for rank, (neighbor_idx, distance) in enumerate(zip(neighbors, distances), 1):
        neighbor_data = df.iloc[neighbor_idx]
        cluster_label = kmeans_labels[neighbor_idx]
        print(f"{rank:<4} {neighbor_idx:<8} {distance:<12.4f} {cluster_label:<8} "
              f"{neighbor_data['age']:<8.2f} {neighbor_data['chol']:<8.2f}")

    print("-" * 60)

# 5. KESIMPULAN
print("\n5. KESIMPULAN DAN REKOMENDASI")
print("-" * 50)
print("ðŸ“Š HASIL ANALISIS:")
print(f"â€¢ Dataset: {df.shape[0]} sampel, {df.shape[1]} fitur")
print(f"â€¢ K-Means: 2 cluster dengan Silhouette {sil_kmeans:.3f} (lebih baik)")
print(f"â€¢ DBSCAN: {n_clusters_dbscan} cluster dengan {n_noise} noise points")
print(f"â€¢ Annoy: Berhasil menemukan nearest neighbors dengan konsisten")

print("\nðŸŽ¯ REKOMENDASI:")
print("â€¢ K-Means dengan k=2 merupakan pendekatan terbaik untuk dataset ini")
print("â€¢ Cluster yang terbentuk menunjukkan pola yang meaningful dalam data medis")
print("â€¢ Annoy efektif untuk pencarian similar patients dalam aplikasi healthcare")

print("\n" + "="*80)
print("ANALISIS SELESAI - MODEL SIAP UNTUK IMPLEMENTASI")
print("="*80)
```
