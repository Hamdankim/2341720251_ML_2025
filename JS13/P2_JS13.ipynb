```python

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import tensorflow as tf
import numpy as np
```

```python
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1,1)

encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python

model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
hist_base = model.fit(X_train, y_train, epochs=50, batch_size=8, verbose=0)
loss_base, acc_base = model.evaluate(X_test, y_test, verbose=0)
acc_base
```

```python

model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(12, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
hist_t2 = model2.fit(X_train, y_train, epochs=50, batch_size=8, verbose=0)
loss_t2, acc_t2 = model2.evaluate(X_test, y_test, verbose=0)
acc_t2
```

```python

model_sig = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='sigmoid', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='sigmoid'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model_sig.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
hist_sig = model_sig.fit(X_train, y_train, epochs=50, batch_size=8, verbose=0)
loss_sig, acc_sig = model_sig.evaluate(X_test, y_test, verbose=0)
acc_sig
```
