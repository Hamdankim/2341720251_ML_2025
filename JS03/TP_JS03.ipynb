```python
#Tugas Praktikum
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# 1. Load data
df = pd.read_csv("/content/wbc (1).csv")

# 2. Drop kolom yang tidak digunakan
df = df.drop(columns=["id", "Unnamed: 32"], errors="ignore")

# 3. Encoding diagnosis (B -> 0, M -> 1)
le = LabelEncoder()
df["diagnosis"] = le.fit_transform(df["diagnosis"])

# 4. Pisahkan fitur (X) dan target (y)
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]

# 5. Standardisasi fitur numerik
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6. Stratified split data (80:20)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# 7. Cek hasil
print("Ukuran data latih :", X_train.shape, y_train.shape)
print("Ukuran data uji   :", X_test.shape, y_test.shape)
print("Distribusi kelas di data latih:\n", y_train.value_counts(normalize=True))
print("Distribusi kelas di data uji:\n", y_test.value_counts(normalize=True))
```
